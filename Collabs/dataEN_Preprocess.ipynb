Convert mp3 to wav

```python
import os
from pydub import AudioSegment
path = "./EN/clips/"

for files in os.listdir(path):
  file_name, file_extension = os.path.splitext(path+files)
  if file_extension == ".mp3":
    sound = AudioSegment.from_mp3(path+files)
    sound.export("{}.wav".format(file_name), format="wav")
    os.remove(path+files)
```

```python
import os
import librosa

holmesPath = "./fr/lupincontresholme/"
lesmisPath = "./fr/lesmis/"

for fileWav in os.listdir(holmesPath):
  duration = librosa.get_duration(filename=(holmesPath+fileWav))
  if duration < limit[0] or duration > limit[1]:
    os.remove(holmesPath+fileWav)

for fileWav in os.listdir(lesmisPath):
  duration = librosa.get_duration(filename=(lesmisPath+fileWav))
  if duration < limit[0] or duration > limit[1]:
    os.remove(lesmisPath+fileWav)
```
